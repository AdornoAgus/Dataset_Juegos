```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from mlxtend.feature_selection import SequentialFeatureSelector as SFS

```

```python
path = r"C:\Users\Admin\Desktop\DATA SCIENCE\DatasetsGame\Video_Games.csv"

DATA = pd.read_csv(path, low_memory=False)
DATA = DATA.dropna(subset=['Year_of_Release'])
DATA['Critic_Score'].fillna(0, inplace=True)
DATA['Year_of_Release'] = pd.to_datetime(DATA['Year_of_Release'], format='%Y', errors='coerce').dt.year
df = DATA[['Name', 'Platform', 'Year_of_Release', 'Genre', 'Publisher', 'NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales', 'Global_Sales', 'Critic_Score']]

df.head(10)
```

```python
X = df[['NA_Sales', 'EU_Sales', 'JP_Sales', 'Other_Sales']]
Y = df['Global_Sales']
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
model_ = LinearRegression()
sfs = SFS(model_,
         k_features=(1, len(X.columns)),
           forward=False,
           floating=False,
           scoring='neg_mean_squared_error',
           cv=5)

sfs.fit(X_train, y_train)
features = list(sfs.k_feature_names_)
print("Features seleccionados: ", features)

model = LinearRegression()
model.fit(X_train[features], y_train)
y_pred = model.predict(X_test[features])
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Error cuadratico medio(RMSE): ", mse)
print("Aptitud del modelo(R^2): ", r2)
```

En conclusion, el modelo nos dio unos buenos resultados de prueba. Utilice columnas que ya sabia estaban correlacionadas, siendo las de ventas segun region y la global.
En cuanto a los resultados, en el error cuadratico medio obtuvimos un resultado bajo, indicando que los puntos de los datos observados estan cerca de los puntos predichos. Mientras que para el R^2, obtuvimos que la aptitud del modelo es casi perfecta con un 0.9999863557674393 de 1.
